```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("D:/Yolo1/runs/detect/train/weights/best.pt")
```

```python
results = model.predict(source="D:/Yolo1/ongvschauchau_chon.mp4", stream=True, conf=0.5)

for result in results:
    frame = result.plot()  # vẽ bounding box
    cv2.imshow("AgroPest Detection (Slow Mode)", frame)
    
    # waitKey(50) -> delay 50ms mỗi frame, tăng lên thì video chạy chậm hơn
    if cv2.waitKey(80) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
